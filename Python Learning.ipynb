```python
def anagram_checker(string1, string2):
    """""""""""
    This sees if two strings are anagrams of one another
    """""""""""
    for char in string1:
        if char in string2:
            string2 = string2.replace(char, '', 1)
        else:
            print(string2)
            return False
    if string2 == '':
        return True
    else:
        return False
```

```python
anagram_checker('rail safety','fairy tales')
```
